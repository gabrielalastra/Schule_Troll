```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path ='/content/drive/MyDrive/School Food 4 Change/Schulen Mittagessenversorgung und Grundinfos.xlsx'
```

```python
df = pd.read_excel(path)
```

```python
df.head()
```

```python
df.rename(columns = {'Unnamed: 2':'Schul-Typ', 'Unnamed: 3':'Ort'}, inplace = True)
df.head()
```

```python
df = df[df.columns[:4]].copy()
```

```python
df
```

```python
plt.figure(figsize=[10,10])
df['Schul-Typ'].value_counts().plot(kind='pie',
                                   colormap='Paired',
                                   autopct='%.2f%%')
my_circle=plt.Circle( (0,0), 0.5, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```

```python
plt.figure(figsize=[15,6])
sns.countplot(x=df['Ort'], data=df)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=[12,6])

ax = sns.countplot(y='Status', data=df, palette='Paired')

total = len(df['Status'])
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))
plt.show()
```

```python
plt.figure(figsize=[12,6])

ax = sns.countplot(y='Schul-Typ', data=df, palette='Paired')
total = len(df['Schul-Typ'])
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))
plt.show()
```
