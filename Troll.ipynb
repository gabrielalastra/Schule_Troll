```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = [
      #['Heidesand', '03-06-22', '-', 0.054],
      ['Mandelhörnchen', '03-06-22', 0.045, 0.052, '10.06.22', 0.042,0.049],
      ['Mandel Schoki', '03-06-22', 0.05, 0.056,'10.06.22',0.05,0.055],
      ['Nougatringe', '03-06-22', 0.056, 0.079, '10.06.22',0.053,0.078],
      ['Nussecke', '03-06-22', 0.036, 0.042,'10.06.22', 0.031,0.036],
      ['Schokotaler', '03-06-22', 0.062, 0.068, '10.06.22',0.058,0.066],
      ['Schweineohr', '03-06-22', 0.062, 0.073,'14.06.22',0.074,0.085],
      ['Schokotörtchen', '03-06-22', 0.05, 0.068,'14.06.22', 0.059,0.084]]
```

```python
df = pd.DataFrame(data=df)
df.head
```

```python
col_names = ['Produkt', 'Datum 01', 'VOR schokolieren 01', 'NACH schokolieren 01', 'Datum 02', 'VOR schokolieren 02', 'NACH schokolieren 02']
df.columns=col_names
df
```

```python
df['Datum 01'] = pd.to_datetime(df['Datum 01'])
df['Datum 02'] = pd.to_datetime(df['Datum 02'])
```

```python
df['VOR schokolieren 01'] = pd.to_numeric(df['VOR schokolieren 01'], errors='coerce').convert_dtypes()
df['NACH schokolieren 01'] = pd.to_numeric(df['NACH schokolieren 01'], errors='coerce').convert_dtypes()
df['VOR schokolieren 02'] = pd.to_numeric(df['VOR schokolieren 02'], errors='coerce').convert_dtypes()
df['NACH schokolieren 02'] = pd.to_numeric(df['NACH schokolieren 02'], errors='coerce').convert_dtypes()
```

```python
df.info()
```

```python
df['Untershied 01'] = (df['NACH schokolieren 01'] - df['VOR schokolieren 01'])
df['Untershied 02'] = (df['NACH schokolieren 02'] - df['VOR schokolieren 02'])
df
```

```python
df.index = df['Produkt']
```

```python
plt.figure(figsize=[30,6])
plt.subplot(1,2,1)
df['Untershied 01'].plot(kind='barh', color='#b3cbf5')
plt.axvline(df['Untershied 01'].mean(), color='red', label='Durchschnitt')
plt.title('Untershied 01')
plt.legend()

plt.subplot(1,2,2)
df['Untershied 02'].plot(kind='barh', color='#b3cbf5')
plt.axvline(df['Untershied 02'].mean(), color='red', label='Durchschnitt')
plt.title('Untershied 02')
plt.legend()
```
